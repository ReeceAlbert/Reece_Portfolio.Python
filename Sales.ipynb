```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
X = sales.drop(columns=['Market Size','Date'])
y = sales['Market Size']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
orig_columns = X_train.columns
lr = LogisticRegression(C=1)
ct = ColumnTransformer([
        ('ohe', OneHotEncoder(), ['State','Market','Product','Type','Product Type'],)
    ], remainder='passthrough')

    
ct.fit(X_train,y_train)
```

```python
X_train_trans = ct.transform(X_train)
```

```python
lr.fit(X_train_trans,y_train)
```

```python
X_test_trans = ct.transform(X_test)
print(lr.score(X_train_trans,y_train))
print(lr.score(X_test_trans,y_test))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
y_pred = lr.predict(X_test_trans)
plt.rcParams['figure.figsize'] = (5, 5)
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=lr.classes_[[0, 1]])
disp.plot(xticks_rotation=90)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train_trans,y_train)
knn.score(X_test_trans,y_test)
```

```python
for i in [1,6,7,8,9,10]:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_trans,y_train)
    print(knn.score(X_test_trans,y_test))
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_trans,y_train)
knn.score(X_test_trans,y_test)
```

```python
y_pred = knn.predict(X_test_trans)
plt.rcParams['figure.figsize'] = (5, 5)
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=knn.classes_[[0, 1]])
disp.plot(xticks_rotation=90)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train_trans,y_train)
dt.score(X_test_trans,y_test)
```

```python
y_pred = dt.predict(X_test_trans)
plt.rcParams['figure.figsize'] = (5, 5)
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=dt.classes_[[0, 1]])
disp.plot(xticks_rotation=90)
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
gb = GradientBoostingClassifier(max_depth=20)
gb.fit(X_train_trans,y_train)
print(gb.score(X_test_trans,y_test))
```

```python
y_pred = gb.predict(X_test_trans)
plt.rcParams['figure.figsize'] = (5, 5)
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=gb.classes_[[0, 1]])
disp.plot(xticks_rotation=90)
```

```python
for i in [1,2,3,4,5,67,8,9,10]:
    rf = RandomForestClassifier(max_depth=i)
    rf.fit(X_train_trans,y_train)
    print(rf.score(X_test_trans,y_test))
```

```python
rf = RandomForestClassifier(max_depth=7)
rf.fit(X_train_trans,y_train)
print(rf.score(X_test_trans,y_test))
```

```python
y_pred = rf.predict(X_test_trans)
plt.rcParams['figure.figsize'] = (5, 5)
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=rf.classes_[[0, 1]])
disp.plot(xticks_rotation=90)
```
